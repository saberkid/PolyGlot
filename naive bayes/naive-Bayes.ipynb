```python
'''
First, construct Alphabet and count the probability of each language

'''
NEW_TRAIN_X = open("data/train_set_x_cleaned_nothing.txt","r")
NEW_TRAIN_Y = open("data/train_set_y.txt","r")

#alphabet for all languages
Alphabet = []
#calculate the probability of each language in the dataset
'''
sl = 0
fr = 1
sp = 2
gr = 3
po = 4
'''
Psl = 0.0
Pfr = 0.0
Psp = 0.0
Pgr = 0.0
Ppo = 0.0

count = 1
for lineY in NEW_TRAIN_Y:
    lang = (lineY.split(","))[1].replace("\n", "")
    text = (NEW_TRAIN_X.readline()).split(",")[1].replace("\n", "")
    chars = text.split()
    if((int)(lang)) == 0:
        Psl = Psl+1
    if((int)(lang)) == 1:
        Pfr = Pfr+1
    if((int)(lang)) == 2:
        Psp = Psp+1
    if((int)(lang)) == 3:
        Pgr = Pgr+1
    if((int)(lang)) == 4:
        Ppo = Ppo+1
    for char in chars:
        if char not in Alphabet:
            Alphabet.append(char)

Sum = Psl+Pfr+Psp+Pgr+Ppo
Psl = Psl/Sum
Pfr = Pfr/Sum
Psp = Psp/Sum
Pgr = Pgr/Sum
Ppo = Ppo/Sum

NEW_TRAIN_X.close()
NEW_TRAIN_Y.close()
```

```python
'''
Second, calculate the w0 and w in the above formula
C0 is the W0 for Slovak and C1 is the W0 for French and so on
w is the weight w in the above representation
'''

import numpy as np
import math as m
import heapq
#Calculate the constant C in boundary 
#C0 is for slovak , C1 is for French etc 
C0 = 0
C1 = 0
C2 = 0
C3 = 0
C4 = 0
w = [[0 for i in range(len(Alphabet))] for j in range(5)]
#constant for slovak
for i in range(len(Alphabet)):
    C0 = C0 + m.log((1-occur[i][0])/(1-(np.sum(occur[i])-occur[i][0])))
    #If there is no occurence of this symbol in the language then we mark its weight as 0
    if(occur[i][0] == 0.0):
        w[0][i] = 0
        continue
    w[0][i] = m.log(occur[i][0] /(np.sum(occur[i])- occur[i][0])) - m.log((1-occur[i][0])/(1-(np.sum(occur[i])-occur[i][0])))
#French 
for i in range(len(Alphabet)):
    C1 = C1 + m.log((1-occur[i][1])/(1-(np.sum(occur[i])-occur[i][1])))
    if(occur[i][1] == 0.0):
        w[1][i] = 0
        continue
    w[1][i] = m.log(occur[i][1] /(np.sum(occur[i])- occur[i][1])) - m.log((1-occur[i][1])/(1-(np.sum(occur[i])-occur[i][1])))
#Spanish
for i in range(len(Alphabet)):
    C2 = C2 + m.log((1-occur[i][2])/(1-(np.sum(occur[i])-occur[i][2])))
    if(occur[i][2] == 0.0):
        w[2][i] = 0
        continue
    w[2][i] = m.log(occur[i][2] /(np.sum(occur[i])- occur[i][2])) - m.log((1-occur[i][2])/(1-(np.sum(occur[i])-occur[i][2])))
for i in range(len(Alphabet)):
    C3 = C3 + m.log((1-occur[i][3])/(1-(np.sum(occur[i])-occur[i][3])))
    if(occur[i][3] == 0.0):
        w[3][i] = 0
        continue
    w[3][i] = m.log(occur[i][3] /(np.sum(occur[i])- occur[i][3])) - m.log((1-occur[i][3])/(1-(np.sum(occur[i])-occur[i][3])))
for i in range(len(Alphabet)):
    C4 = C4 + m.log((1-occur[i][4])/(1-(np.sum(occur[i])-occur[i][4])))
    if(occur[i][4] == 0.0):
        w[4][i] = 0
        continue
    w[4][i] = m.log(occur[i][4] /(np.sum(occur[i])- occur[i][4])) - m.log((1-occur[i][4])/(1-(np.sum(occur[i])-occur[i][4])))

C0 = C0+m.log(Psl/(1-Psl))
C1 = C1+m.log(Pfr/(1-Pfr))
C2 = C2+m.log(Psp/(1-Psp))
C3 = C3+m.log(Pgr/(1-Pgr))
C4 = C4+m.log(Ppo/(1-Ppo))
```

```python
'''
Here the Accuracy from the Validation set is being calculated
'''
#Here is the section for prediction
validation = open("data/validation.csv","r")
fout = open("data/mistakes.csv","a")
success = 0
fail = 0
total = 0
for test in validation:
    lang = (test.split(","))[0]
    letters = ((test.split(","))[1].replace("\n", "")).split(" ")
    decisions = [C0,C1,C2,C3,C4]
    prediction = 0
    for letter in letters:
        try:
            index = Alphabet.index(letter)
        except:
            #assume non-seen character probability is 0.5
            decisions[i] = decisions[i] + 0.5
            continue
        for i in range(5):
            decisions[i] = decisions[i] + w[i][index]
    prediction = int(decisions.index(max(decisions)))
    if( test.split(",")[1].replace("\n", "") == '' ):
        prediction = 3
    if( prediction == int(lang)):
        success = success + 1
        total = total + 1
    else:
        fout.write('prediction is  ' + str(prediction) + "  but real answer is   " + lang + '\n')
        fout.write(test)
        fout.write('boundaries are ' + str(decisions).strip('[]') + '\n' + '\n')
        fail = fail + 1
        total = total + 1
print (success*1.0/total)
```
